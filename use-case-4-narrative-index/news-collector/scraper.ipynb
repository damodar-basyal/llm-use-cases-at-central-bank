```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

# URL of the Money section
url = "https://kathmandupost.com/money"

# Send a GET request to fetch the HTML content
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find all article elements
    articles = soup.find_all('article', class_='article-image')

    # Initialize a list to store extracted news
    news_list = []

    # Extract details from each article
    for article in articles:
        title = article.find('h3').text.strip() if article.find('h3') else None
        link = article.find('a')['href'] if article.find('a')['href'] else None
        author_tag = article.find('span', class_='article-author')
        author = author_tag.text.replace('By ', '').strip() if author_tag else 'Unknown'
        content = article.find('p').text.strip() if article.find('p') else None
        
        print(f"Title: {title}")
        print(f"Link: https://kathmandupost.com{link}")
        print(f"Author: {author}")
        print(f"Content: {content}\n")

        # Add the article details to the list
        if title and content:
            news_list.append({
                "title": title,
                "link": link,
                "description": content,
                "author": author
            })
```

```python
news_list
```

```python
#Saving data
today = datetime.now().strftime("%Y-%m-%d")

if not os.path.exists(f'../data/News/{today}'):
    os.mkdir(f'../data/News/{today}')

# Save the data as a JSON file
with open(f'../data/News/{today}/KathmanduPost.json', 'w') as file:
    json.dump(news_list, file, indent=4)
```
